# A short introduction to Classes in Python

The ability to define and manipulate Classes in Python is the backbone of Object Oriented Programming (OOP) in Python. Without further adieu, let's jump straight into how Classes work with examples.

## The very basics

```python
# Creating a simple class called Dog
# It has an instance attribute called breed. This is an instance attribute because it depends on the specific instance of Dog.
class Dog:
    def __init__(self,breed):  # The __init__ method is special: It is defines and sets initial values of class and instance attributes
        self.breed = breed   # "self" is a reference to the instance of the class
    
    def bark(self): # This is an example of a method. It is a function that can called by all instances of this Class to do something.
        print('Woof woof.')
```

We have created a simple class called `Dog`. Just as in the real world, there can be many instances of dogs. Let's define two instances of `Dog`.

```python
dog_1 = Dog(breed = 'Golden Retriever')
dog_2 = Dog(breed = 'Border Collie')

# We have not written a method to retrieve the instance attribute 'breed', but we may as well just obtain it directly like this
print(f'The first dog\'s breed is: {dog_1.breed}.\nThe second dog\'s breed is: {dog_2.breed}.') 

dog_1.bark() # Just to show that these methods work and how one calls them for each instance.
dog_2.bark()
```

## Inheritance

One of the most useful features of classes is called _Inheritance_. Essentially, one can define subclasses of classes which inherit the attributes and methods of the parent class (or superclass). Below is an example of this.

```python
import math
# Note: All numerical outputs have been formatted to return 3 significant figures. It's just cleaner that way! :)

class Shape:
    def __init__(self, dimensions):
        self.dimensions = dimensions
    
    def how_many_dims(self):
        print(f'This shape is {self.dimensions}-dimensional.')

# Below is how you create a subclass -- same syntax except you name the parent class in parentheses.
class Circle(Shape): # Notice how "dimensions" is not initialised as a Circle attribute. 
                     # Instead, it is permanantly set to be 2 using the parent class's attribute.
    def __init__(self, radius):
        super().__init__(dimensions = 2)
        self.radius = radius

    def area(self):
        return f'{math.pi * self.radius**2:.3g}'

    def perimeter(self):
        return f'{2 * math.pi * self.radius:.3g}'
    

class Rectangle(Shape):
    def __init__(self, length, breadth):
        super().__init__(dimensions = 2)
        self.length = length # Just as with Circle, the Rectangle subclass has its own instance attributes. 
        self.breadth = breadth # In this cases, Rectangle has 'length' and 'breadth' but no 'radius'.
    
    # We may define the notion of area and perimeter for rectangles too, but these are obviously different from those for circles!
    def area(self):
        return f'{self.length * self.breadth:.3g}'
    
    def perimeter(self):
        return f'{2 * (self.length + self.breadth):.3g}'
    

class Sphere(Shape):
    def __init__(self, radius):
        super().__init__(dimensions = 3)
        self.radius = radius

    # Spheres have volume and surface area instead of area and perimeter.
    def volume(self):
        return f'{4/3 * math.pi * self.radius**3:.3g}'
    
    def surface_area(self):
        return f'{4 * math.pi * self.radius**2:.3g}'

my_circle = Circle(radius = 1) # Create an instnace of a Circle with unit radius.
my_circle.how_many_dims() # The Circle subclass has access to the how_many_dims() method of the parent!
print(f'The area of my_circle is: {my_circle.area()} and the perimeter of my_circle is: {my_circle.perimeter()}.') 

my_rectangle = Rectangle(length = 4, breadth = 3)
print(f'The area of my_rectangle is: {my_rectangle.area()} and the perimeter of my_rectangle is: {my_rectangle.perimeter()}.') 

smol_sphere = Sphere(radius = 2)
chonk_sphere = Sphere(radius = 10)

print(f'The volumes and surface areas of the smol_sphere and chonk_sphere are {smol_sphere.volume()}, {smol_sphere.surface_area()}\
 and {chonk_sphere.volume()}, {chonk_sphere.surface_area()} respectively.')

# As expected, Circle, Rectangle, and Sphere all have access to the attribute "dimensions" 
# and the method "how_many_dims" from the Shape parent class.
# Simultaneously, they each have their own attributes and methods that Shape does not have.
# The "area" and "perimeter" methods of Circle and Rectangle differ in functionality.
```

Feel free to play around with more stuff with the above subclasses! The key takeaway is that subclasses inherit parent methods and attributes. This is an incredibly powerful feature. Subclasses themselves can contain their own methods and attributes that may be unique to (instances of) them. Below is another example highlighting a minor subtlety of inheritance.

```python
class Square(Rectangle): 
    def __init__(self, length): # First, you don't need to make every parent attribute something that is user-defined.
                                # In this case, expecting the user to duplicate their "length" input for "breadth" is a waste of time and effort.

        super().__init__(length = length, breadth = length) # Instead, we solicit one input ("length") for an instance of Square 
                                                            # and assign that to be equal to the breadth attribute from the Rectangle class

my_square = Square(4)

# Reiterating that the Square subclass inherits everything from Rectangle and Shape.
my_square.how_many_dims()
print(f'Area: {my_square.area()}, Perimeter = {my_square.perimeter()}.')
```

Notice how sparse the code for setting up Square is. And yet, any square we create has a length, breadth, dimensions = 2. AND, we can compute the area and perimeter of every instance of Square. Hopefully, this conveys how powerful inheritance chains can be and how easy it is to set them up neatly!

## Naming conventions for attributes and methods

One underscore is used in front of attributes and methods to indicate to the user that they are not meant to be accessed directly. It isn't an issue to just print it out, but it may be a problem if th developer/whoever wrote the code decides to remove that attribute or method and you've build something that relies on calling it explicitly. 

```python
class Parallelogram():
    def __init__(self, length, breadth, acute_angle):
        self.length = length
        self.breadth = breadth
        self.acute_angle = acute_angle

    def _calculate_height(self):
        return self.breadth * math.sin(self.acute_angle)
    
    def area(self):
        height = self._calculate_height()
        return height * self.length
    

my_par = Parallelogram(length = 6, breadth = 4, acute_angle=math.pi/6)

print(f'The height of this parallelogram is {my_par._calculate_height():.3g}.')
print(f'The area of this parallelogram is {my_par.area():.3g}')

# Here is what NOT to do (not just because it's mathematically stupid, 
# but also because the hidden method could be removed from the implementation of Parallelogram entirely.

height = my_par._calculate_height()
other_angle = math.pi/2 + math.acos(height/my_par.breadth)

print(f'The obtuse angle of my parallelogram is {other_angle:.3g} in radians or {other_angle * 180/math.pi:.3g} degrees.')
```

## The Method Resolution Order (MRO)


Finally, we present some subtleties on the method resolution order (MRO). The MRO is the order that Python follows to look up attributes and methods in a class hierarchy. Take the example below. `A` is a parent of both `B` and `C`, and `D` is a subclass/child of both `B` and `C`. As we mentioned earlier, `D` inherits "everything" from `A`,`B`, and `C`. But what if `B` and `C` have the same method/attribute? Which one is preferred? Run the code below to see!

```python
class A:
    pass

class B(A):
    def hello(self):
        print("Hello from B")

class C(A):
    def hello(self):
        print("Hello from C")

class D(B, C):
    pass

obj = D()
print(D.mro()) # The method resolution order (MRO) of a subclass can be obtained like this. Saves a lot of headache.

obj.hello()
```

Indeed, Python uses the MRO of Class `D` to resolve this ambiguity. As can be seen from the output of `print(D.mro())`, Python first checks `D` itself, and then `B`, followed by `C`, and finally `A`. Since `D` does not directly have a method called `hello()`, Python checks `B` next. Since `B` does have this method, Python executes it accordingly and stops the chain of checking there, since the method has been executed.

Next, let us take a look at a more subtle example which elucidates how the method `super()` works. Consider the setup of four classes below.

Observe what happens below when we take out the `super().method()` from class C. 

```python
class A:
    def method(self):
        print("A method")


class B(A):
    def method(self):
        print("B method")
        super().method()


class C(A):
    def method(self):
        print("C method")
        # super().method()


class D(B, C):
    def method(self):
        print("D method")
        super().method()


obj = D()
print(D.mro())
# print(obj.mro())

obj.method()
```

Does it makes sense in relation to the MRO of class D? At first glance, you might think that once `method()` is called from `D` and Python moves to Class `B` to call `method()` again, the `super().method()` in `B` should then tell Python to go to the superclass of `B` according to its MRO, which would lead us to `method()` in Class `A`. However, this is clearly not what happens. 

What `super()` does, regardless of where it is called, is to check the MRO of the _initiating_ object's class not the MRO of the _current_ class where the executing method is being called. Hence, when the code reaches `method()` in Class `B`, it executes `print("B method")` and then `super().method()`. The latter tells Python to the MRO of the class of `obj`, go to the next Class in that MRO, and then look for the method called `method()`. Python therefore realises that it is curently in Class `B` of the MRO chain `[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]` and accordingly moves to check Class `C` for a `method()` method. Since it is there, Python executes it, so we see "C method" printed. However, since there is no `super().method()` in Class `C`, Python stops moving further down the MRO chain of the class of `obj` and so the code terminates.
